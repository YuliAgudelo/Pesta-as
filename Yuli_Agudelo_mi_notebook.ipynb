```python
import pandas as pd
def_salario = pd.read_csv('salary_data.csv', sep=';')
def_salario
```

```python
el_nivel_educacion_es_2 = def_salario['education_level']== 2
df_filtrado = def_salario[el_nivel_educacion_es_2]
df_filtrado
```

```python
max_sueldo = df_filtrado['income'].max()
print(f'El máximo sueldo es: ${max_sueldo}')
```

```python
min_sueldo = df_filtrado['income'].min()
print(f'El mínimo sueldo es: ${min_sueldo}')
```

```python
sueldo_medio = df_filtrado['income'].mean()
print(f'El sueldo medio es: ${sueldo_medio}')
```

```python
sueldo_mediano = df_filtrado['income'].median()
print(f'El sueldo mediano es: ${sueldo_mediano}')
```

```python
# Rango de edades
edad_minima = df_filtrado['age'].min()
edad_maxima = df_filtrado['age'].max()
print(f'El rango de edades es de {edad_minima} a {edad_maxima} años')
```
